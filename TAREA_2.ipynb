```python
# Importar las librerías necesarias
import numpy as np
from scipy.integrate import quad
from scipy.special import roots_legendre

# Definición de funciones
def f1(x):
    return (1 / np.sqrt(2 * np.pi)) * np.exp(-x**2 / 2)

def f2(x):
    return (np.exp(x) * np.sin(x)) / (1 + x**2)

def f3(x):  # Función para el punto 4
    return np.exp(x) / x

def g(x):  # Función para el punto 4
    return (1 - np.exp(x)) / x

# Método del trapecio
def trapecio(func, a, b, n):
    x = np.linspace(a, b, n+1)
    y = func(x)
    h = (b - a) / n
    return h * (np.sum(y) - 0.5 * (y[0] + y[-1]))

# Método de Simpson
def simpson(func, a, b, n):
    if n % 2 != 0:
        n += 1
    x = np.linspace(a, b, n+1)
    y = func(x)
    h = (b - a) / n
    return h / 3 * (y[0] + y[-1] + 4 * np.sum(y[1:-1:2]) + 2 * np.sum(y[2:-2:2]))

# Método de Gauss-Legendre
def gauss_legendre(func, a, b, n):
    [x, w] = roots_legendre(n)

    t = 0.5 * (x + 1) * (b - a) + a
    return 0.5 * (b - a) * np.sum(w * func(t))


integrals = [
    {"func": f1, "a": -1, "b": 1, "name": "Integral 1"},
    {"func": f2, "a": 0, "b": 3, "name": "Integral 2"},
    {"func": f2, "a": -2, "b": 0, "name": "Integral 2 (límites cambiados)"},
]

# Adicional para el punto 4
functions_p4 = [
    {"func": f3, "a": 0.1, "b": 5, "name": "f(x) = e^x / x"},
    {"func": g, "a": 0.1, "b": 5, "name": "g(x) = (1 - e^x) / x"},
]


partitions = [6, 15, 20]
points_gauss = [6, 15]

# Calcular las integrales
for integral in integrals:
    func = integral["func"]
    a, b = integral["a"], integral["b"]
    print(f"\n{integral['name']} en [{a}, {b}]:")

    for n in partitions:
        result_trap = trapecio(func, a, b, n)
        result_simp = simpson(func, a, b, n)
        print(f"  Trapecio (n={n}): {result_trap:.8f}")
        print(f"  Simpson (n={n}): {result_simp:.8f}")

    for n in points_gauss:
        result_gauss = gauss_legendre(func, a, b, n)
        print(f"  Gauss-Legendre (n={n}): {result_gauss:.8f}")

    # Valor exacto usando scipy
    exact, _ = quad(func, a, b)
    print(f"  Valor exacto (scipy): {exact:.8f}")

# Punto 4: Integración de f(x) y g(x)
print("\nPunto 4: Integración de funciones f(x) y g(x):")
for function in functions_p4:
    func = function["func"]
    a, b = function["a"], function["b"]
    print(f"\n{function['name']} en [{a}, {b}]:")

    for n in partitions:
        result_trap = trapecio(func, a, b, n)
        result_simp = simpson(func, a, b, n)
        print(f"  Trapecio (n={n}): {result_trap:.8f}")
        print(f"  Simpson (n={n}): {result_simp:.8f}")

    for n in points_gauss:
        result_gauss = gauss_legendre(func, a, b, n)
        print(f"  Gauss-Legendre (n={n}): {result_gauss:.8f}")

    # Valor exacto usando scipy
    exact, _ = quad(func, a, b)
    print(f"  Valor exacto (scipy): {exact:.8f}")
```
