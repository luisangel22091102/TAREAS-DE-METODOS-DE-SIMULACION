```python
import math
import numpy as np
import matplotlib.pyplot as plt

# --- Parte A: Probabilidad de 3 mujeres en un grupo de 40 ---

# Número de mujeres y total de alumnos
mujeres = 9
total_alumnos = 40

# Guardar los resultados
probabilidades = []

# Repetir 100 veces el experimento
for _ in range(100):
    combinaciones_mujeres = math.comb(mujeres, 3)
    combinaciones_totales = math.comb(total_alumnos, 3)
    probabilidad = combinaciones_mujeres / combinaciones_totales
    probabilidades.append(probabilidad)

# --- Parte B: Calificaciones aleatorias ---

# Repetir el experimento 100 veces
promedios = []
varianzas = []

for _ in range(100):
    # Generar 100 calificaciones aleatorias entre 5 y 10
    calificaciones = np.random.randint(5, 11, size=100)

    # Calcular el promedio y la varianza
    promedios.append(np.mean(calificaciones))
    varianzas.append(np.var(calificaciones))
# --- Parte C: Promedio y desviación estándar de los resultados de los promedios de 100 experimentos ---

# Calcular promedio y desviación estándar de los promedios obtenidos
promedio_promedios = np.mean(promedios)
desviacion_estandar_promedios = np.std(promedios)

# Calcular promedio y desviación estándar de las varianzas obtenidas
promedio_varianzas = np.mean(varianzas)
desviacion_estandar_varianzas = np.std(varianzas)

# Mostrar los resultados
print(f"Promedio de promedios: {promedio_promedios}")
print(f"Desviación estándar de promedios: {desviacion_estandar_promedios}")
print(f"Promedio de varianzas: {promedio_varianzas}")
print(f"Desviación estándar de varianzas: {desviacion_estandar_varianzas}")

# --- Graficar resultados ---

# Graficar los promedios y varianzas de las 100 repeticiones
plt.figure(figsize=(10, 6))

# Gráfico de los promedios
plt.subplot(2, 1, 1)
plt.plot(promedios, label="Promedios", color='blue')
plt.title("Promedios de Calificaciones - 100 Repeticiones")
plt.xlabel("Repetición")
plt.ylabel("Promedio")
plt.legend()

# Gráfico de las varianzas
plt.subplot(2, 1, 2)
plt.plot(varianzas, label="Varianzas", color='red')
plt.title("Varianzas de Calificaciones - 100 Repeticiones")
plt.xlabel("Repetición")
plt.ylabel("Varianza")
plt.legend()

# Mostrar gráficos
plt.tight_layout()
plt.show()
```

```python
import math
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# --- Parte A: Probabilidad de 3 mujeres en un grupo de 40 ---

# Número de mujeres y total de alumnos
mujeres = 9
total_alumnos = 40

# Guardar los resultados
probabilidades = []

# Repetir 100 veces el experimento
for _ in range(100):
    combinaciones_mujeres = math.comb(mujeres, 3)
    combinaciones_totales = math.comb(total_alumnos, 3)
    probabilidad = combinaciones_mujeres / combinaciones_totales
    probabilidades.append(probabilidad)

# --- Parte B: Calificaciones aleatorias ---

# Repetir el experimento 100 veces
promedios = []
varianzas = []

for _ in range(100):
    # Generar 100 calificaciones aleatorias entre 5 y 10
    calificaciones = np.random.randint(5, 11, size=100)

    # Calcular el promedio y la varianza
    promedios.append(np.mean(calificaciones))
    varianzas.append(np.var(calificaciones))
# --- Parte C: Promedio y desviación estándar de los resultados de los promedios de 100 experimentos ---

# Calcular promedio y desviación estándar de los promedios obtenidos
promedio_promedios = np.mean(promedios)
desviacion_estandar_promedios = np.std(promedios)

# Calcular promedio y desviación estándar de las varianzas obtenidas
promedio_varianzas = np.mean(varianzas)
desviacion_estandar_varianzas = np.std(varianzas)

# Mostrar los resultados
print(f"Promedio de promedios: {promedio_promedios}")
print(f"Desviación estándar de promedios: {desviacion_estandar_promedios}")
print(f"Promedio de varianzas: {promedio_varianzas}")
print(f"Desviación estándar de varianzas: {desviacion_estandar_varianzas}")

# --- Graficar resultados: Densidad de probabilidad de los promedios ---

plt.figure(figsize=(10, 6))

# Graficar la densidad de probabilidad de los promedios
sns.kdeplot(promedios, shade=True, color='blue', label="Densidad de Probabilidad de Promedios")
plt.title("Densidad de Probabilidad de los Promedios de Calificaciones - 100 Repeticiones")
plt.xlabel("Promedio")
plt.ylabel("Densidad de Probabilidad")
plt.legend()

# Mostrar gráfico
plt.show()


```
