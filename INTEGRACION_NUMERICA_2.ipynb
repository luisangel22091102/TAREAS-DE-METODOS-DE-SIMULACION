```python
%matplotlib inline
import numpy as np
from sympy import*
x=Symbol ('x')

f= exp(4*x)
plot (f,(x,0,1))
```

```python
# Inciso a)
from sympy import integrate
from sympy.abc import x
funcion = exp(4*x)
i = integrate(funcion, x)
print(i)
```

```python
# Inciso a)
def f(x):
    return math.exp(4*x)
n = 100
a = 0
b = 1
dx = (b-a)/n
suma = 0
for i in range(n):
    area = (f(a) + f(a+dx))*dx/2
    suma = suma + area
    a = a + dx
print ("el area es", suma)
```
